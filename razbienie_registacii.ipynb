```python
import openpyxl
import shutil
from openpyxl.styles import Alignment
import numpy as np
from google.colab import files
import os
import calendar
import sys

# from xlsx2html import xlsx2html
# import pdfkit



def month_name (number):
    if number == 1:
        return "Январь"
    elif number == 2:
        return "Февраль"
    elif number == 3:
        return "Март"
    elif number == 4:
        return "Апрель"
    elif number == 5:
        return "Май"
    elif number == 6:
        return "Июнь"
    elif number == 7:
        return "Июль"
    elif number == 8:
        return "Август"
    elif number == 9:
        return "Сентябрь"
    elif number == 10:
        return "Октябрь"
    elif number == 11:
        return "Ноябрь"
    elif number == 12:
        return "Декабрь"

wb = openpyxl.load_workbook('Учет результатов.xlsx')#,keep_vba=True)
sheetlist = wb.sheetnames[1:18]#[1:19]
i = 1
for sheet in sheetlist:
  ws = wb[sheet]
  # print(sheet)
 
  list_for_del = []
  for row in ws.iter_rows(min_row=2, min_col=2, max_col=5):
    pp=[]
    # pp.append(row[0].row)
    for cell in row:
      pp.append(cell.value)
    # print(pp)
    if all((el is None)or(len(el) < 4) for el in pp):
      list_for_del.append(row[0].row)
  
  for num in reversed(list_for_del):
    # ws.delete_rows(num)
    ws.row_dimensions[num].hidden = True
wb.save('Учет результатов-2.xlsx')
wb.close()
# sys.exit(0)
        
file_name = str(month_name(12))+'-2021.xlsx'
shutil.copyfile('Учет результатов-2.xlsx',file_name)
wb = openpyxl.load_workbook(file_name,keep_vba=True)

wrap_alignment = Alignment(wrap_text=True)
sheetlist = wb.sheetnames[1:19]

for i in reversed(range(2, 13)):
    file_name = str(month_name(i)) + '-2021.xlsx'
    wb = openpyxl.load_workbook(file_name)#,keep_vba=True)
    colll = 0
    for sheet in sheetlist:
        list1 = []
        start_delete = 0
        ws = wb[sheet]
        for col_cells in ws.iter_cols(min_row=2, min_col=2, max_col=2):
            for cell in col_cells:
                if cell.value is not None: 
                  p_month = cell.value[2:4]              
                  if(p_month.isnumeric()):
                    if int(p_month) == i:
                        start_delete = cell.row
                        list1.append(cell.row)

        a = np.array(list1)
        if len(a)==0:
            # sys.exit(0)
            continue
        colll += 1
        print(f'{colll} - {a[0]} - {ws}')

        end_row = a[0] + 1

        print('ws.max_row = ',ws.max_row)
        # for roww in reversed(ws[f'A{a[0]-1}:W{ws.max_row}']):
        #   for cell in roww:
        #     cell.value = None
          # ws.row_dimensions[index_roww].hidden = True
          # print(type(roww))

        ws.delete_rows(a[0]-1, len(list(ws.rows))+1)
        # for row_hid in reversed(range(end_row,ws.max_row)):
        #   ws.row_dimensions[row_hid].hidden = True
        

        
        # p_cell = ws.cell(row = end_row, column = 17)
        # p_cell = ws[f"R{end_row}"]
        p_cell = ws[f"R{end_row}"]
        # print('--------------------------',ws.max_row,end_row)
        p_cell.alignment = wrap_alignment
        if i<=6:
          p_cell.value = f'Землякова В.В.               Проведен мониторинг записей (формул){calendar.monthrange(2021, i-1)[1]}.{str(i-1).zfill(2)}.2021г. Булгакова Е.Н.'
        else:
          p_cell.value = f'Землякова В.В.               Проведен мониторинг записей (формул){calendar.monthrange(2021, i-1)[1]}.{str(i-1).zfill(2)}.2021г. Боченкова О.А.'
        ws.row_dimensions[end_row].height = 90
        print_area = f"A1:W{end_row}"
        ws.print_area = print_area
        ws.page_setup.orientation = ws.ORIENTATION_LANDSCAPE
        ws.sheet_view.view = "pageBreakPreview"
        ws.sheet_properties.pageSetUpPr.fitToPage = True
        ws.page_setup.fitToHeight = False

    # wb.save(file_name)#---------

    print(f'{month_name(i - 1)} done....')
    file_name = str(month_name(i - 1)) + '-2021.xlsx'
    wb.save(file_name)
    wb.close()

    # files.download(file_name)
    # os.remove(missing_ok=True)


```
